```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data=np.array([[85.0,117.0],
    [83.5,110.0],
    [78.2,102.0],
    [68.1,97.0],
    [66.3,86.0]])

instance=data.shape[0]
print(data)
```

```python
plt.figure(figsize=(10,5))
plt.xlabel("CS")
plt.ylabel("GO")
plt.scatter(x=data[:,0], y=data[:,1])
plt.title("CS:GO")
plt.show()
```

```python
def SSE(m,b,data):
    total_error=0.0
    
    for i in range(instance):
        x=data[:,0]
        y=data[:,1]
        
        current_rating=y
        
        predict_rating=m*x+b
        
        error= (current_rating-predict_rating)**2
        
        total_error+=error
    
    sse=total_error/instance
    print (sse)
    return(sse)
        
```

```python
m=3
b=2
sse=SSE(m,b,data)

```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
ax.set_title('ADR vs Rating (CS:GO)')
ax.scatter(x=data[:,0],y=data[:,1],label='Data')
plt.plot(data[:,0], m*data[:,0] + b,color='red',label='Our Fitting Line')
ax.set_xlabel('ADR')
ax.set_ylabel('Rating')
ax.legend(loc='best')

plt.show()
```
